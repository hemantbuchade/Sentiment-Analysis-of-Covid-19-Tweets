# **Data Processing and Cleaning**

# Removing User

```python
df['OriginalTweet'].iloc[4117]
```

```python
def rem_user(tweet):
  a=[]
  for word in tweet.lower().split():
    if word.startswith('@'):
      continue
    else:
      a.append(word)
  return ' '.join(a)
```

```python
df['Tweets'] =df['OriginalTweet'].apply(rem_user) 
```

```python
df['Tweets'].iloc[4117]
```

# Removing links

```python
def rem_links(tweet):
  a=[]
  for word in tweet.lower().split():
    if word.startswith('https:') or word.startswith('http:'):
      continue
    else:
      a.append(word)
  return ' '.join(a)
```

```python
df['Tweets'] =df['Tweets'].apply(rem_links) 
```

```python
df['Tweets'].iloc[4117]
```

# Removing digits and special characters

```python
import re
def rm_dig(tweet):
  a=[]
  for word in tweet.split():
    output = re.sub('[â£/0-9]+', '', word)
    a.append(output)
  return ' '.join(a)
```

```python
df['Tweets'] =df['Tweets'].apply(rm_dig) 
```

```python
df['Tweets'].iloc[4117]
```

# Removing stopwords and Punctuations

```python
import nltk
nltk.download('stopwords')
```

```python
import string
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
[punc for punc in string.punctuation]
```

```python
def text_process(msg):
    nopunc =[char for char in msg if char not in string.punctuation]
    nopunc=''.join(nopunc)
    return ' '.join([word for word in nopunc.split() if word.lower() not in stopwords.words('english')])
```

```python
df['Tweets'] = df['Tweets'].apply(text_process)
```

```python
df['Tweets'].iloc[4117]
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer

stemmer = PorterStemmer()
def stem_words(text):
     return " ".join([stemmer.stem(word) for word in text.split()])

df['Tweets']=df['Tweets'].apply(stem_words)
```

```python
df['Tweets'].iloc[4117]
```

# Lemmatization

```python
lemmatizer = WordNetLemmatizer()
wordnet_map = {"N":wordnet.NOUN, "V":wordnet.VERB, "J":wordnet.ADJ, "R":wordnet.ADV}
def lemmatize_words(text):
    pos_tagged_text = nltk.pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_tagged_text])
```

```python
df['Tweets']=df['Tweets'].apply(lemmatize_words)
```

# Analyzing Hashtags

```python
def get_tags(tweet):
  stri = ''
  for word in tweet.lower().split():
    if word.startswith("#"):
      stri+=word[1:]+','
  return stri[:-1]
```

```python
df['hashtags'] = df['OriginalTweet'].apply(get_tags)
```
